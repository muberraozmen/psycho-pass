```python
import pandas as pd
from pyrit.datasets import SeedDatasetProvider

dataset_names = SeedDatasetProvider.get_all_dataset_names()
dataset_names = [d for d in dataset_names if d.startswith("airt_") or d in ["adv_bench", "harmbench"]]

stats = []
for d in dataset_names:
    try:
        dataset = await SeedDatasetProvider.fetch_datasets_async(dataset_names=[d])
        seeds = dataset[0].seeds
        stats.append({
            "dataset_name": d,
            "description": dataset[0].description,
            "author": dataset[0].author,
            "num_seeds": len(seeds),
        })
    except Exception as e:
        continue

stats = pd.DataFrame(stats)

stats
 
```
