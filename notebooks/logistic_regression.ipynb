```python
import os 
os.chdir("/Users/muberraozmen/Development/psycho-pass")

import warnings
warnings.filterwarnings('ignore', category=RuntimeWarning)

from pathlib import Path
import numpy as np
import pandas as pd
from src.factory import ClassifierFactory

run_dir = Path("./experiments/objective_model_comparisons/")

def show_results(args):
    classifier_factory = ClassifierFactory(args, run_dir)
    data, performance, factors = classifier_factory.execute()
    print("Configuration:")
    for k, v in args.items():
        print(f"  {k} = {v}")
    print("\n")
    print(f"Dataset Statistics:\n  Average Success Rate: {data['outcome'].mean():.2f}\n  Number of Samples: {len(data)}\n")
    print(f"Classification Performance: \n{performance}\n")
    print(f"Significant Factors: \n{factors}\n")
    
    
```
